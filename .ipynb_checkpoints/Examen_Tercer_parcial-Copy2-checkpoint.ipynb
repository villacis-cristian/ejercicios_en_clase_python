```python
import pandas as pd  # Para manipulación y análisis de datos
import numpy as np   # Para operaciones numéricas
import matplotlib.pyplot as plt  # Para visualización de datos
import seaborn as sns  # Para gráficos estadísticos más avanzados
from sklearn.model_selection import train_test_split  # Para dividir datos en entrenamiento y prueba
from sklearn.linear_model import LinearRegression  # Modelo de regresión lineal
from sklearn.metrics import mean_squared_error, r2_score  # Métricas de evaluación
from sklearn.preprocessing import StandardScaler  # Para estandarización de datos
import matplotlib.pyplot as plt
```

# EXAMEN FINAL TERCER PARCIAL VIDA O MUERTE TODO SI NO NADA MUERTE AL PUNK

```python
# Cargar el dataset desde el archivo CSV
df = pd.read_csv('movie_genre_classification_final.csv')  

# Mostrar información general del dataset
print(df.info())  # Ver tipos de datos y valores nulos
print(df.describe())  # Estadísticas descriptivas
```

# EXAMEN FINAL TERCER PARCIAL VIDA O MUERTE TODO SI NO NADA MUERTE AL PUNK

```python
# Seleccionar variables numéricas relevantes
features_reg = ['Budget_USD', 'Duration', 'Rating', 'Votes', 'Num_Awards', 'Critic_Reviews']  
target_reg = 'BoxOffice_USD'  

# Filtrar solo las columnas necesarias
data_reg = df[features_reg + [target_reg]].copy()  

# Eliminar filas con valores nulos
data_reg.dropna(inplace=True)  

# Verificar correlación entre variables
corr_matrix = data_reg.corr()  
plt.figure(figsize=(10, 8))  
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')  
plt.title("Correlación entre Variables")  
plt.show()  
```

# EXAMEN FINAL TERCER PARCIAL VIDA O MUERTE TODO SI NO NADA MUERTE AL PUNK

```python
# Dividir en conjuntos de entrenamiento (80%) y prueba (20%)
X_reg = data_reg[features_reg]  
y_reg = data_reg[target_reg]  
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(  
    X_reg, y_reg, test_size=0.2, random_state=42  
)  

# Estandarizar los datos (importante para regresión lineal)
scaler_reg = StandardScaler()  
X_train_reg_scaled = scaler_reg.fit_transform(X_train_reg)  
X_test_reg_scaled = scaler_reg.transform(X_test_reg)  
```

# EXAMEN FINAL TERCER PARCIAL VIDA O MUERTE TODO SI NO NADA MUERTE AL PUNK

```python
# Crear y entrenar el modelo
model_reg = LinearRegression()  
model_reg.fit(X_train_reg_scaled, y_train_reg)  

# Predecir en los conjuntos de entrenamiento y prueba
y_train_pred_reg = model_reg.predict(X_train_reg_scaled)  
y_test_pred_reg = model_reg.predict(X_test_reg_scaled)  

# Calcular métricas de rendimiento
mse_train = mean_squared_error(y_train_reg, y_train_pred_reg)  
mse_test = mean_squared_error(y_test_reg, y_test_pred_reg)  
r2_train = r2_score(y_train_reg, y_train_pred_reg)  
r2_test = r2_score(y_test_reg, y_test_pred_reg)  

print(f"MSE (Train): {mse_train}")  
print(f"MSE (Test): {mse_test}")  
print(f"R² (Train): {r2_train}")  
print(f"R² (Test): {r2_test}")  
```

# EXAMEN FINAL TERCER PARCIAL VIDA O MUERTE TODO SI NO NADA MUERTE AL PUNK

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12, 8))

# Calcular residuos
residuals = y_test_reg - y_test_pred_reg

# Scatter plot con tamaño según magnitud del residuo
sns.scatterplot(x=y_test_pred_reg, y=residuals,
                hue=np.abs(residuals),
                size=np.abs(residuals),
                palette='coolwarm',
                sizes=(20, 200),
                alpha=0.7,
                edgecolor='w',
                linewidth=0.5)

# Líneas de referencia
plt.axhline(y=0, color='r', linestyle='--', label='Residuo Cero')
plt.axhline(y=residuals.mean(), color='green', 
            linestyle=':', label=f'Media (${residuals.mean():,.0f})')

# Añadir KDE plot para visualizar densidad
sns.kdeplot(x=y_test_pred_reg, y=residuals,
            cmap='Blues', 
            fill=True,
            alpha=0.3)

# Calcular estadísticas de residuos
residual_std = np.std(residuals)
within_1std = np.mean(np.abs(residuals) < residual_std) * 100

# Añadir texto con estadísticas
stats_text = f"""Estadísticas de Residuos:
Media = ${residuals.mean():,.0f}
Desviación Estándar = ${residual_std:,.0f}
{within_1std:.1f}% dentro de ±1σ
Máximo error = ${np.max(np.abs(residuals)):,.0f}"""

plt.text(0.02, 0.95, stats_text, 
         transform=plt.gca().transAxes,
         bbox=dict(facecolor='white', alpha=0.8),
         fontsize=10,
         verticalalignment='top')

# Configuraciones adicionales
plt.xlabel('Valores Predichos (USD)', fontsize=12)
plt.ylabel('Residuos (Real - Predicho)', fontsize=12)
plt.title('Análisis de Residuos del Modelo de Regresión', 
          fontsize=14, pad=20)
plt.grid(True, linestyle='--', alpha=0.3)
plt.legend(loc='upper right')
plt.tight_layout()

# Remover leyenda de tamaño/color (opcional)
current_handles, current_labels = plt.gca().get_legend_handles_labels()
plt.legend(current_handles[:3], current_labels[:3])

plt.show()
```
